## Glitch categorization

```python
def label_single_glitches(df):
    df = df.copy()
    mask = (df['response'].notna()) & (df['response'].isin(DUMMY_FREE_ISOGENY_NEIGHBOURS.keys()))
    df.loc[mask,'expected_degree'] = df.loc[mask, 'response'].apply(lambda x: DUMMY_FREE_ISOGENY_NEIGHBOURS[int(x)][1])
    df.loc[mask,'expected_degree_str'] = df.loc[mask, 'response'].apply(lambda x: DUMMY_FREE_ISOGENY_NEIGHBOURS[int(x)][0])
    df.loc[mask,'expected_absolute_degree'] = df.loc[mask, 'response'].apply(lambda x: abs(DUMMY_FREE_ISOGENY_NEIGHBOURS[int(x)][1]))
    
    df['interval'] = -1  # default (no match)
    df['actual_degree'] = None
    df['actual_absolute_degree'] = None
    
    
    for i, entry in enumerate(APPROX_EXECUTION_TIME.to_dict('records')):
        interval_start = entry['setup_start']
        interval_end = entry['isogeny_end'] if not i == len(APPROX_EXECUTION_TIME) - 1 else APPROX_TOTAL_EXECUTION_TIME 
        actual_degree = entry['degree']
    
        # Assign interval ID where offset falls in range
        mask = (interval_start < df['ext_offset']) & (df['ext_offset'] < interval_end)
        df.loc[mask, 'interval'] = i
        df.loc[mask, 'actual_degree'] = actual_degree
        df.loc[mask, 'actual_orientation'] = 1 if actual_degree > 0 else -1
        df.loc[mask, 'actual_absolute_degree'] = abs(actual_degree)
    
        mask = ()
    
    df = df.dropna(subset=["interval"])
    len(df), len(df[df.type.isin(['JUSTRIGHT']) & df.expected_degree.notna()])
    return df

df = label_single_glitches(data['s'])
```

```python
def filter_faulty_outputs(df):
    df = df[df.expected_degree.notna()].copy()
    df = df.sort_values("ext_offset").reset_index(drop=True)
    threshold = 0
    dominance_factor = 1
    print(len(df))
    keep = [True]
    for i in range(1, len(df)):
        if df.loc[i, "ext_offset"] - df.loc[i - 1, "ext_offset"] < threshold:
            keep.append(False)
        else:
            keep.append(True)
    df = df[keep].reset_index(drop=True)
    print(len(df))
    return df
```

```python
def compare_counts(results, dominance_factor):
    P = results["(+)"] + results["(++)"]
    N = results["(-)"] + results["(--)"]

    if P > dominance_factor * N:
        return +1
    elif N > dominance_factor * P:
        return -1
    else:
        return 0
```

```python
def guess_orientations(groups_per_degree, dominance_factor=1, correct_col=True):
    correct = 0
    table = []
    for d, groups in groups_per_degree.items():
    
        # print("=" * 30)
        for group in groups:
            results = {
                "(+)": len(group[group.expected_degree == d]),
                "(++)": len(group[group.expected_degree == d**2]),
                "(-)": len(group[group.expected_degree == -d]),
                "(--)": len(group[group.expected_degree == -d**2])
            }
            pattern = compare_counts(results, dominance_factor=dominance_factor)
    
            if pattern > 0:
                p = '+'
            elif pattern < 0:
                p = '-'
            else:
                p = '?'
    
            if pattern > 0 and group.iloc[0].actual_degree > 0:
                correct += 1
    
            if pattern < 0 and group.iloc[0].actual_degree < 0:
                correct += 1
    
            degree = group.iloc[0].actual_degree
            entry = {
                "Isogeny step": group.iloc[0].interval,
                "Degree": degree,
                "Group size": len(group[group.expected_degree_str.notna()]),
                "Dominating orientation": f"${p}$",
            }
            if correct_col:
                entry["Correct"] =  "$\\cmark$" if (pattern > 0 and degree > 0) or (pattern < 0 and degree < 0) else "-"

            entry.update(results)
            table.append(entry)   
    print(f"Correctly guessed orientations: {correct}.")
    table = pd.DataFrame(table)
    return table, correct
```

```python
def label_double_glitches(data, degree):
    DUMMY_FREE_ISOGENY_NEIGHBOURS= {
        3: {
    
            368: ("(-)3", -3),
            404: ("(--)3", -3**2),
            75:  ("(---)3", -3**3),
            144: ("(----)3", -3**4),
            158: ("(+)3", 3), 
            0:   ("(++)3", +3**2),
            261: ("(+++)3", +3**3),
            9:   ("(++++)3", +3**4),
            410: ("(P)", 0)
        },
        5: {
            40:  ("(+)5", +5),
            75:  ("(++)5", +5**2),
            228: ("(+++)5", +5**3),
            174: ("(++++)5", +5**4),
            29:  ("(-)5", -5),
            261: ("(--)5", -5**2),
            124: ("(---)5", -5**3),
            15:  ("(----)5", -5**4),
            410: ("(P)", 0)
        },
        7: {
            51:  ("(+)7", +7),
            245: ("(++)7", +7**2),
            29:  ("(+++)7", +7**3),
            413: ("(++++)7", +7**4),
    
            191: ("(-)7", -7),
            199: ("(--)7", -7**2),
            40:  ("(---)7", -7**3),
            344: ("(----)7", -7**4),
            410: ("(P)", 0)
        }
    }
    
    DUMMY_FREE_ISOGENY_NEIGHBOURS = DUMMY_FREE_ISOGENY_NEIGHBOURS[degree]
    df = data.copy()
    # df = df[df.type != "CHANGING"]
    # df = df[df.type == 'JUSTRIGHT']
    df['ext_offset_0'] = df['ext_offset'].apply(lambda x: x[0])
    df['ext_offset_1'] = df['ext_offset'].apply(lambda x: x[0]+x[1])
    # We are making sure that both offsets were found to be correctly skipping the isogeny 
    # df = df[df['ext_offset_0'].isin(POSSIBLY_SKIPPED.ext_offset) & df['ext_offset_1'].isin(POSSIBLY_SKIPPED.ext_offset)]
    df['interval_0'] = None
    df['interval_1'] = None
    
    mask = (df['response'].isin(DUMMY_FREE_ISOGENY_NEIGHBOURS.keys()))
    df.loc[:, 'expected_degree'] = None
    df.loc[mask,'expected_degree'] = df.loc[mask, 'response'].apply(lambda x: DUMMY_FREE_ISOGENY_NEIGHBOURS[int(x)][1])
    
    cat = sorted([x[0] for x in DUMMY_FREE_ISOGENY_NEIGHBOURS.values()])
    print(cat)
    df.loc[mask,'expected_degree_str'] = df.loc[mask, 'response'].apply(lambda x: DUMMY_FREE_ISOGENY_NEIGHBOURS[int(x)][0])
    df.loc[mask, 'expected_degree_str'] = pd.Categorical(df.loc[mask, 'expected_degree_str'], cat)
    df['actual_absolute_degree_0'] = None
    df['actual_absolute_degree_1'] = None
    
    
    for i, entry in enumerate(APPROX_EXECUTION_TIME.to_dict('records')):
        interval_start = entry['setup_start']
        interval_end = entry['isogeny_end'] if not i == len(APPROX_EXECUTION_TIME) - 1 else APPROX_TOTAL_EXECUTION_TIME 
        actual_degree = entry['degree']
        actual_absolute_degree = abs(entry['degree'])
    
        # Assign interval ID where offset falls in range
        mask_0 = (interval_start < df['ext_offset_0']) & (df['ext_offset_0'] < interval_end)
        mask_1 = (interval_start < df['ext_offset_1']) & (df['ext_offset_1'] < interval_end)
    
        df.loc[mask_0, 'interval_0'] = i
        df.loc[mask_1, 'interval_1'] = i
    
    
        df.loc[mask_0, 'actual_degree_0'] = actual_degree
        df.loc[mask_1, 'actual_degree_1'] = actual_degree
        
        df.loc[mask_0, 'actual_absolute_degree_0'] = actual_absolute_degree
        df.loc[mask_1, 'actual_absolute_degree_1'] = actual_absolute_degree
    
    
        mask_0 = ()
        mask_1 = ()
    print(len(df))
    return df
```

```python
import pandas as pd
def build_table(df, degree, desired):
    g = df[df.expected_degree_str.notna()].copy()
    
    # Extract orientation symbol including "(P)"
    g["orientation"] = g.expected_degree_str.str.extract(
        r"(\(+\+*\)|\(-+\)|\(P\))"
    )
    
    # Count occurrences
    summary = (
        g.groupby(
            ["interval_0", "interval_1", "actual_degree_0", "actual_degree_1", "orientation"]
        )["unit"]
         .count()
         .reset_index()
    )
    
    # Pivot so orientations become columns
    pivot = summary.pivot_table(
        index=["interval_0", "interval_1", "actual_degree_0", "actual_degree_1"],
        columns="orientation",
        values="unit",
        fill_value=0
    ).reset_index()

    # Ensure all orientation columns exist
    for col in desired:
        if col not in pivot:
            pivot[col] = 0

    # Reorder columns
    pivot = pivot[
        ["interval_0", "interval_1", "actual_degree_0", "actual_degree_1"] + desired
    ]

    # Convert to int
    for c in pivot.columns:
        pivot[c] = pivot[c].astype(int)

    return pivot
```
